```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load DB
df = pd.read_csv("customer_data.csv")

df.head()

# Drop cust_id column
df_clean = df.drop('cust_id', axis=1)

# Check and fill missing values
df_clean.isnull().sum()
df_clean.fillna(df_clean.mean(), inplace=True)

# Standardize features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_clean)

# KMeans Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(df_scaled)

df['cluster'] = clusters
```

```python
# PCA reduction
# Reduce to 2D for visualization
pca = PCA(n_components=2)
pca_data = pca.fit_transform(df_scaled)

plt.figure(figsize=(8,6))
plt.scatter(pca_data[:,0], pca_data[:,1], alpha=0.5)
plt.title('PCA Projection (Unlabeled)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
# Elbow Method
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Plot clusters in 2D PCA space
plt.figure(figsize=(8,6))
sns.scatterplot(x=pca_data[:,0], y=pca_data[:,1], hue=clusters, palette='Set1')
plt.title('Customer Segmentation with KMeans')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Cluster')
plt.show()
```

```python
# Analyze each cluster
numeric_cols = df.select_dtypes(include=np.number).columns

cluster_profile = df[numeric_cols].groupby(df['cluster']).mean()
cluster_profile
```
